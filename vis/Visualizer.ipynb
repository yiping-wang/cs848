```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import animation
from matplotlib.animation import FuncAnimation
```

```python
def animate(i, subj, top_k, palette):
    df = pd.read_csv("./{}/universities_{}_{}_{}.csv".format(subj, subj, 2010 + i, 2010 + i))
    unis = list(df.head(top_k)['Uni'])
    counts = list(df.head(top_k)['count'])
    plt.barh(range(top_k), counts, color=palette)
    plt.title("{} Top {} in Year {} ".format(subj, top_k, 2010 + i), color=("blue"))
    plt.yticks(np.arange(top_k), unis, fontsize=7, rotation=45)
```

```python
subj = 'vision'
top_k = 10

palette = list(reversed(sns.color_palette("Spectral", top_k).as_hex()))[::-1]
fig = plt.figure(figsize=(10,5))
axes = fig.add_subplot(1,1,1)
axes.set_xlim(0, 15)
plt.style.use("seaborn")
axes.set_xlabel("{}".format(subj))
ani = FuncAnimation(fig, animate, frames=12, interval=1500, fargs=(subj, top_k, palette))
plt.show()
```
